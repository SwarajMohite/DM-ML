## make a new column for episode count

```python
def ep_count(title):
    ep_count = ""
    flag = False
    for i in title:
        if i ==")":
            flag = False
            return ep_count
        if flag == True :
            ep_count += i
        if i == "(":
            flag = True


df['Episode count'] = df['Title'].apply(ep_count).str.replace(" eps","").astype(int)

df['Episode count'].info()
```

## make a new column for time stamp

```python
def timestamp(title):
    count = 0
    flag = False
    timestamp = ""
    for j in range(21):
        for i in title:
            if flag == True:
                timestamp += i
                count += 1
            if i == ")":
                flag = True
            if count >= 19 :
                return timestamp

df['Timestamp'] = df['Title'].apply(timestamp)
```

## which is the longest running anime

```python
from datetime import datetime

def month(Timestamp):
    start = ""
    end = ""
    start, end = Timestamp.split(" - ")
    start_date = datetime.strptime(start, '%b %Y')  
    end_date = datetime.strptime(end, '%b %Y')

    total = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1

    return total

df['Month'] = df['Timestamp'].apply(month)

df.head()
```

```python
long_run_anime = df[df['Month'] == df['Month'].max()]
long_run_anime
```
