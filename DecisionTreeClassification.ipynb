```python
import pandas as pd
df = pd.read_csv("Product_Dataset.csv")

df
```

```python
input = df[['Spiciness','Sourness']]

output = df['Class']

print(input)
print(output)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()

tree.fit(input,output)
```

```python
tamarind = pd.DataFrame({'Spiciness' : [4], 'Sourness':[10]})
tamarind

```

```python
tree.predict(tamarind)
```

```python
Cayenne =  pd.DataFrame({'Spiciness' : [5], 'Sourness':[1]})
tree.predict(Cayenne)
```

```python
from sklearn.tree import plot_tree

plot_tree(tree);
```
